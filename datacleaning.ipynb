```python
import sqlite3

conn = sqlite3.connect("hospital.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS patients (
    PatientID VARCHAR,
    Name TEXT,
    Age INTEGER,
    Gender TEXT,
    VisitDate VARCHAR,
    Cholesterol INTEGER,
    SugarLevel INTEGER,
    diagnosis TEXT
);
""")

conn.commit()
```

```python
Conn = sqlite3.connect("hospital.db")
Patient_Data.to_sql("patients", conn, if_exists="replace", index=False)
conn.close()

print("Data inserted successfully!")
```
